```python
columns_to_concat = ['title', 'genres', 'keywords', 'original_language', 'overview', 'tagline', 'cast', 'director']
df['combined_column'] = df[columns_to_concat].astype(str).apply(lambda row: ' '.join(row), axis=1)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  text = str(text).lower()
  text = re.sub(r'[^a-zA-Z\s]', '', text)
  words = text.split()
  words = [word for word in words if word not in stop_words]
  words = [lemmatizer.lemmatize(word) for word in words]
  return ' '.join(words)
```

```python
df['combined_column'] = df['combined_column'].apply(preprocess_text)
```

```python
df.sample(5)
```

```python
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
```

```python
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english', max_features= 50000, ngram_range=(1,2))
```

```python
tfidf_matrix = tfidf.fit_transform(df['combined_column'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n= 10):
  if title not in indices:
    return "Title not found"

  idx = indices[title]

  print(f"Title : {df['title'].iloc[idx]}")
  print(f"Director : {df['director'].iloc[idx]}")
  print(f"Cast : {df['cast'].iloc[idx]}")
  print(f"Genres : {df['genres'].iloc[idx]}")

  sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_scores.argsort()[::-1][1:n+1]

  results = pd.DataFrame({
      "title": df['title'].iloc[similar_idx].values,
      "similarity_score": sim_scores[similar_idx],
      "Cast": df['cast'].iloc[similar_idx].values,
      "Director": df['director'].iloc[similar_idx].values,
      "Genres": df['genres'].iloc[similar_idx].values,
  })
  return results
```
